# Minimum Cost Spanning Tree
Problem Description:
A spanning tree of a graph can be defined as a graph with minimal set of edges that connect all vertices. A minimum spanning tree of a graph is a spanning tree of the graph with least weight (where the weight is computed by adding the weights of all the edges in the spanning tree). In general, a graph can have multiple minimum spanning trees. The problem is to find a minimum spanning tree of a graph.

This implementation uses a dictionary to represent the weighted graph, where the vertices are keys, and the values are dictionaries containing their adjacent vertices and their weights.

Please note that this implementation assumes that the input graph is connected and weighted.

First, let's create a simple graph to demonstrate Prim's algorithm:

```python
graph = {
    "A": {"B": 10, "C": 15, "D": 20},
    "B": {"A": 10, "C": 12},
    "C": {"A": 15, "B": 12, "D": 18},
    "D": {"A": 20, "C": 18},
}
```

Now, let's implement Prim's algorithm:

```python
import heapq

def prim_algorithm(graph):
    if not graph:
        return [], 0
    start_vertex = list(graph.keys())[0]
    mst_vertices = set([start_vertex])


    mst_edges = []
    edges_to_consider = []
    for neighbor, weight in graph[start_vertex].items():
        heapq.heappush(edges_to_consider, (weight, start_vertex, neighbor))

    min_total_weight = 0

    while len(mst_vertices) < len(graph):
        if not edges_to_consider:
            break

        weight, u, v = heapq.heappop(edges_to_consider)

        if v in mst_vertices:
            continue

        mst_vertices.add(v)
        mst_edges.append((u, v, weight))
        min_total_weight += weight

        for neighbor, new_weight in graph[v].items():
            if neighbor not in mst_vertices:
                heapq.heappush(edges_to_consider, (new_weight, v, neighbor))

    return mst_edges, min_total_weight
```

Now, let's run the Prim's algorithm on the example graph:
Keep in mind that this example uses a simple representation of a graph, but for more complex cases, you might want to use a library such as NetworkX to represent and manipulate graphs.

```python
mst, min_weight = prim_algorithm(graph)
print("Minimum Spanning Tree: ", mst)
print("Minimum Weight: ", min_weight)
```

Use same apporach for Kruskal's algorithm

```python
def find_parent(parent, i):
    if parent[i] == i:
        return i
    parent[i] = find_parent(parent, parent[i])
    return parent[i]

def union_sets(parent, rank, x, y):
    x_root = find_parent(parent, x)
    y_root = find_parent(parent, y)
    if x_root != y_root:
        if rank[x_root] < rank[y_root]:
            parent[x_root] = y_root
        elif rank[x_root] > rank[y_root]:
            parent[y_root] = x_root
        else:
            parent[y_root] = x_root
            rank[x_root] += 1
        return True
    return False

def kruskal_algorithm(graph):
    edges = []
    for u in graph:
        for v, weight in graph[u].items():
            # Add each edge only once (e.g., from u to v, but not v to u)
            if (v, u, weight) not in edges:
                edges.append((u, v, weight))

    # Sort edges by weight
    edges.sort(key=lambda x: x[2])

    vertices = list(graph.keys())
    parent = {v: v for v in vertices}
    rank = {v: 0 for v in vertices}

    mst = []
    min_weight = 0
    num_edges = 0

    for u, v, weight in edges:
        if num_edges == len(vertices) - 1:
            break

        if union_sets(parent, rank, u, v):
            mst.append((u, v, weight))
            min_weight += weight
            num_edges += 1

    return mst, min_weight

# Run Kruskal's algorithm on the lecture graph
mst_kruskal, min_weight_kruskal = kruskal_algorithm(graph_lecture)
print("Kruskal's Minimum Spanning Tree: ", mst_kruskal)
print("Kruskal's Minimum Weight: ", min_weight_kruskal)
```
